```python
import numpy as np
import pandas as pd
import seaborn as sns
import PIL
from PIL import ImageEnhance
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib import image    
```

```python
# convert horse images into flatten 
a=[]
for i in range(1,1047):
    
    t=Image.open(r"C:\Users\mounika katla\Downloads\Horses_Augmented-20230506T132537Z-001\Horses_Augmented\horse ({}).jpg".format(i))
    y=ImageEnhance.Color(t)
    y=y.enhance(0)
    y=np.array(y)
    z=y.flatten()
    a.append(z)
df=pd.DataFrame(a)
df
    
```

```python
# labelling the horse images with 0
df["label"]=0
```

```python
# converting donkey images into flatten 
b=[]
for i in range(1,971):
    
    t=Image.open(r"C:\Users\mounika katla\Downloads\Donkeys_Augmented-20230506T132534Z-001\Donkeys_Augmented\Donkey ({}).jpg".format(i))
    y=ImageEnhance.Color(t)
    y=y.enhance(0)
    y=np.array(y)
    z=y.flatten()
    b.append(z)
df2=pd.DataFrame(b)
df2
    
```

```python
# labelling the donkey with 1
df2["label"]=1
```

```python
# concatting the horse and donkey's dataframe
final_df=pd.concat([df,df2],axis=0)
```

```python
# separating predictable and target columns
y=final_df["label"]
X=final_df.drop(columns="label")
```

```python
# spliting data into train and test
from sklearn.model_selection import train_test_split
xTrain,xTest,yTrain,yTest=train_test_split(X,y,
                                        test_size=0.20,
                                        random_state=23)
```

```python
xTrain.shape,xTest.shape,len(yTrain),len(yTest)
```

#  creating the models

# buliding the decision tree model

```python
import sklearn
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT=DecisionTreeClassifier() # object

DT.fit(xTrain,yTrain)
```

```python
yTrainPred,yTestPred=DT.predict(xTrain),DT.predict(xTest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(yTrain,yTrainPred),accuracy_score(yTest,yTestPred)
```
